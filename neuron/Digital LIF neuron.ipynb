# Notebook description

This notebook looks at the digital implementation of the LIF neuron. The dynamics of the continuous time LIF neuron are described by

$$\tau\dot{x}=-x+u$$

where 
 - $x$ is the neuron state
 - $\tau$ is the membrane time constant and
 - and $u$ is the input to the neuron.
 
When $x$ reaches a predefined threshold $x_{t}$, we declare the neuron as having spiked, reset $x$ to its reset value $x_t$ (typically $0$), and fix $x$ for a refactory period $t_{ref}$. 

This notebook looks at the digital LIF neuron. The digital LIF neuron differs from the continuous LIF neuron in that the digital neuron integrates on a $\Delta t$ time step and instead of resetting $x$ to $0$, subtracts the threshold from $x$. That is, the digital neuron resets $x$ to $x-x_t$. Given these changes, we would like to know what is the fi curve for the digital neuron.

# LIF neuron analysis

The output of the LIF neuron is spikes. We first derive the spike rate of an LIF neuron assuming that the input is constant. We do this by first finding the time, $t_t$ it takes for the neuron to reach threshold and then note that substitute into the equation describing the firing rate 

$$f=\frac{1}{t_{ref}+t_t}$$

Assuming that $x=x_0$ at $t=0$, 

\begin{align}
\tau\frac{dx}{dt} &= -x + u \\
\frac{\tau}{-x+u}dx &= dt \\
\int_{x_0}^{x}\frac{\tau}{-x+u}dx &= \int_{0}^{t} dt \\
\tau\int_{x_0}^{x}\frac{1}{-x+u}dx &= \int_{0}^{t} dt. \\
\end{align}

Substituting with $v=-x+u$ (so $-dv=dx$),

\begin{align}
\tau\int_{x_0}^{x}\frac{1}{-x+u}dx &= \int_{0}^{t} dt \\
-\tau\int\frac{1}{v}dv &= \int_{0}^{t} dt \\
-\tau[\ln(v)]_{v_0}^{v} &= t \\
-\tau[\ln(-x+u)]_{x_0}^{x} &= t \\
-\tau[\ln(-x+u)-\ln(-x_0+u)] &= t \\
-\tau\ln\left(\frac{-x+u}{-x_0+u}\right) &= t. \\
\end{align}

Setting $t=t_t$ when $x=x_t$, we have

$$
-\tau\ln\left(\frac{-x_t+u}{-x_0+u}\right) = t_t,
$$

so the firing rate for an LIF neuron with constant input will be

$$
f = \frac{1}{t_{ref}-\tau\ln\left(\frac{u-x_t}{u-x_0}\right)}.
$$

If we set $x_0=0$, we have the firing rate function typically seen in the literature,

$$
f = \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u}\right)},
$$

However, we'll use the form with the arbitrary initial condition when analyzing the digital LIF neuron.

Solving for $x$ in terms of $t$, 

\begin{align}
-\tau\ln\left(\frac{-x+u}{-x_0+u}\right) &= t \\
\frac{-x+u}{-x_0+u} &= e^{-t/\tau} \\
-x+u &= (-x_0+u)e^{-t/\tau} \\
x &= -(-x_0+u)e^{-t/\tau} + u \\
x &= x_0e^{-t/\tau}+(1-e^{-t/\tau})u. \\
\end{align}

# Deterministic Digital LIF neuron analysis

Discretizing $\dot{x}$, we have the update equation

\begin{align}
x[n] &= e^{-\Delta t/\tau}x[n-1] + (1-e^{-\Delta t/\tau})u[n] \\
 &= \alpha x[n-1] + (1-\alpha)u[n]
\end{align}

where $\alpha = e^{-\Delta t/\tau}$.

To find the firing rate given a constant input, we first find the overshoot.

Start by assuming from $x[0]=x_0$.

\begin{align}
x[0] &= x_0 \\
x[1] &= \alpha x_0 + (1-\alpha)u \\
x[2] &= \alpha^2 x_0 + \alpha(1-\alpha)u + (1-\alpha)u \\
x[3] &= \alpha^3 x_0 + \alpha^2(1-\alpha)u + \alpha(1-\alpha)u + (1-\alpha)u \\
&\vdots \\
x[n] &= \alpha^n x_0 + \sum_{i=0}^{n-1}\alpha^i(1-\alpha)u, \\
\end{align}

so

\begin{align}
x[n] &= \alpha^n x_0 + \frac{1-\alpha^n}{1-\alpha}(1-\alpha)u \\
 &= \alpha^n x_0 + (1-\alpha^n)u \\
 &= \alpha^n(x_0-u) + u. \\
\end{align}

By assuming the update equation is continuous, we can find the step at which we cross threshold. Substituting $x_t$ and solving for $n$,

\begin{align}
x_t &= \alpha^t(x_0 - u) + u \\
x_t - u &= \alpha^t(x_0 - u) \\
\frac{x_t - u}{x_0 - u} &= \alpha^t \\
\log_\alpha\left(\frac{x_t - u}{x_0 - u}\right) &= t \\
\end{align}

The time step at which we pass threshold will be $n=\lceil t\rceil$. Therefore

\begin{align}
\text{overshoot} &= x[n]-x_t \\
 &= \alpha^{\left\lceil\log_\alpha\left(\frac{x_t - u}{x_0 - u}\right)\right\rceil}(x_0-u)+u - x_t\\
\end{align}

## Steady state analysis

At each spike the overshoot will be slightly different. However, we can analyze the steady state case. At steady state, the overshoot will match the initial condition of the neuron. To find the steady state firing rate, we first find $k_T$, the interspike interval in units of $\Delta t$.

\begin{align}
\alpha^{k_T}(x_0-u) + u - x_t &= x_0 \\
\alpha^{k_T}(x_0-u) &= x_0 - u  + x_t\\
\alpha^{k_T} &= \frac{x_0 - u + x_t}{x_0-u} \\
\alpha^{k_T} &= 1+\frac{x_t}{x_0-u} \\
k_T &= \log_\alpha\left(1+\frac{x_t}{x_0-u}\right) \\
k_T &= \frac{1}{\ln(\alpha)}\ln\left(1+\frac{x_t}{x_0-u}\right) \\
k_T &= -\frac{\tau}{\Delta t}\ln\left(1+\frac{x_t}{x_0-u}\right) \\
\end{align}

Given a refractory period $k_{ref}$ also in units of $\Delta t$, the firing rate will be

\begin{align}
f &= \frac{1}{(k_{ref}+k_T)\Delta t} \\
 &= \frac{1}{k_{ref}\Delta t-\tau\ln\left(1+\frac{x_t}{x_0-u}\right)},
\end{align}

which is similar to the continuous LIF neuron form

\begin{align}
f_{\text{cLIF}} = \frac{1}{t_{ref}-\tau\ln\left(\frac{u-x_t}{u-x_0}\right)}.
\end{align}


If I know my desired spike rate in units of $1/\Delta t$, I can determine $x0-u$. If, say

$$f=\frac{1}{(k_{ref}+k_{target})\Delta t},$$

then 

\begin{align}
k_{target} &= -\frac{\tau}{\Delta t}\ln\left(1+\frac{x_t}{x_0-u}\right) \\
-k_{target}\frac{\Delta t}{\tau} &= \ln\left(1+\frac{x_t}{x_0-u}\right) \\
e^{-k_{target}\Delta t/\tau} &= 1+\frac{x_t}{x_0-u} \\
e^{-k_{target}\Delta t/\tau}-1 &= \frac{x_t}{x_0-u} \\
x_0-u &= \frac{x_t}{e^{-k_{target}\Delta t/\tau}-1} \\
\end{align}

```python
class DigitalNeuron(object):
    def __init__(self, dt=.001, tau=.01, kref=1, xt=1., x0=0.):
        self.dt = dt
        self.tau = tau
        self.kref = kref
        self.xt = xt
        self.x0 = x0
        
        self.x = x0
        self.ref = False
        self.ref_ctr = 0
        self.a = np.exp(-dt/tau)
        
    def reset(self):
        self.x = self.x0
        self.ref = False
        self.ref_ctr = 0
        
    def step(self, u):
        assert isinstance(u, (float, int))
        if self.ref:
            self.ref_ctr += 1
            if self.ref_ctr == self.kref:
                self.ref_ctr = 0
                self.ref = False
        else:
            self.x = self.a*self.x + (1-self.a)*u
            if self.x >= self.xt:
                self._begin_refractory_period()
                self.ref = True
                return 1
        return 0
    
    def _begin_refractory_period(self):
        self.x -= self.xt

class DigitalNeuronReset0(DigitalNeuron):
    def __init__(self, dt=.001, tau=.01, kref=1, xt=1., x0=0.):
        super(DigitalNeuronReset0, self).__init__(dt=dt, tau=tau, kref=kref, xt=xt, x0=x0)
        
    def _begin_refractory_period(self):
        self.x = 0
        
def theoretical_neuron(x0, u, n, dt=.001, tau=.01):
    a = np.exp(-dt/tau)
    return a**n*(x0 - u) + u

def ss_diff(k, dt=.001, tau=.01, xt=1.):
    """returns x0-u"""
    return xt/(np.exp(-k*dt/tau)-1)
```

```python
# relation between u and x0 for fixed firing rate

fig = figure(figsize=(8,6))
ax = fig.add_subplot(111)

dt = .001
kt = 5
nsteps = 8*kt
x0_u_ss = ss_diff(kt)

u = 2.6
x0 = x0_u_ss + u
x = np.zeros(nsteps+1)
x[0] = x0
t = np.arange(nsteps+1)*dt
neuron = DigitalNeuron(dt, x0=x0)
for k in xrange(nsteps):
    spk = neuron.step(u)
    x[k+1]=neuron.x
ax.plot(t, x, '-o', label='u=%.1f\nx0=%.2f' % (u, x0))
ax.axhline(y=x0, color='k')

u = 2.7
x0 = x0_u_ss + u
x = np.zeros(nsteps+1)
x[0] = x0
t = np.arange(nsteps+1)*dt
neuron = DigitalNeuron(dt, x0=x0)
for k in xrange(nsteps):
    spk = neuron.step(u)
    x[k+1]=neuron.x
ax.plot(t, x, '-o', label='u=%.1f\nx0=%.2f' % (u, x0))
ax.axhline(y=x0, color='k')

ax.axhline(y=1., color='r')
ax.set_xlabel('time (s)')
ax.set_ylabel('neuron state')
ax.set_title('x0 and u to achieve a fixed firing rate are not unique')
ax.legend(loc='right');
```

```python
def fi_experiment(u, dt=.001, tau=.01, xt=1., neuron_class=DigitalNeuron):
    """empirically measure the fi curve"""
    rates = []
    idx = u>xt
    for u_val in u[idx]:
        spk_times = []
        neuron = neuron_class(dt=dt, tau=tau, xt=xt)
        k=0
        while len(spk_times) < 20 and len(np.unique(np.diff(spk_times[-10:])))!=1:
            spk = neuron.step(u_val)
            if spk:
                spk_times.append(k)
            k += 1
        rate = 1./np.mean(np.diff(spk_times[-10:]))
        rates.append(rate)
    f = np.zeros(u.shape)
    f[idx] = np.array(rates)/dt
    return f
```

```python
# fi curve for threshold subtraction refractory mechanism
fig = figure(figsize=(16,6))

dt = .001
tau = .01
xt = 1.

a = np.exp(-dt/tau)
umax = xt/(1-a)
u = np.logspace(0, np.log10(umax), 100)
f = fi_experiment(u=u, dt=dt, tau=tau, xt=xt, neuron_class=DigitalNeuron)

ax = fig.add_subplot(121)
ax.plot(u, f, '-o')
ax.set_ylabel('f')
ax.set_xlabel('u')
ax = fig.add_subplot(122)
#ax.semilogy(u, f, '-o')
ax.plot(u, 1./(dt*f), '-o')
ax.set_ylabel('period (dt)')
ax.set_xlabel('u')
fig.suptitle('empirically measured fi with threshold subtraction reset is quantized');
```

```python
# can I get a closed form solution to the tuning curve?

def f_digital(u, x0, dt=.001, tau=.01, kref=1, xt=1.):
    return (kref*dt-tau*np.log(1.+xt/(x0-u)))**-1

def f_cont(u, x0=0., tau=.01, tref=.001, xt=1.):
    return (tref-tau*np.log((u-xt)/(u-x0)))**-1.

x0s = np.linspace(0, xt, 4)
u = np.linspace(0, 100.*xt, 51)
idx = u>xt

figure(figsize=(16,6))
subplot(121)
for x0 in x0s:
    f = np.zeros(u.shape)
    f[idx] = f_digital(u[idx], x0, xt=xt)
    plot(u, f, label='%.2f' % x0)
legend(loc='best')

subplot(122)
f[idx] = f_cont(u[idx])
plot(u, f)
```
