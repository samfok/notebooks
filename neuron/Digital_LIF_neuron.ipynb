# Notebook description

This notebook looks at the digital implementation of the LIF neuron. The dynamics of the continuous time LIF neuron are described by

$$\tau\dot{x}=-x+u$$

where 
 - $x$ is the neuron state
 - $\tau$ is the membrane time constant and
 - and $u$ is the input to the neuron.
 
When $x$ reaches a predefined threshold $x_{t}$, we declare the neuron as having spiked, reset $x$ to its reset value $x_t$ (typically $0$), and fix $x$ for a refactory period $t_{ref}$. 

The digital LIF neuron differs from the continuous LIF neuron in that the digital neuron integrates on a $\Delta t$ time step. Instead of resetting $x$ to $0$, the digitial neuron could subtract the threshold from $x$. That is, the digital neuron could reset $x$ to $x-x_t$ or it could resetting $x$ to $0$ depending on the implementation. 

We would like to know the spike rate as a function of the input (i.e. the tuning curve or fi curve) for the digital neuron.

# Digital LIF neuron analysis

Discretizing $\dot{x}$ produces the update equation

\begin{align}
x[n] &= e^{-\Delta t/\tau}x[n-1] + (1-e^{-\Delta t/\tau})u[n] \\
 &= \alpha x[n-1] + (1-\alpha)u[n]
\end{align}

where $\alpha = e^{-\Delta t/\tau}$. The firing rate will be

$$
f = \frac{1}{(k_{ref}+k_t)\Delta t}
$$

where $k_{ref}$ is the refractory period in units of $\Delta t$ and $k_t$ is the period in units of $\Delta t$.

## Digital LIF neuron with reset to 0 refractory mechanism

If we reset $x$ to $0$ upon each spike, we can find the $f$ as a function of $u$.

\begin{align}
x[0] &= 0 \\
x[1] &= (1-\alpha)u \\
x[2] &= \alpha(1-\alpha)u + (1-\alpha)u \\
x[3] &= \alpha^2(1-\alpha)u + \alpha(1-\alpha)u + (1-\alpha)u \\
&\vdots \\
x[n] &= \sum_{i=0}^{n-1}\alpha^i(1-\alpha)u, \\
\end{align}

so

\begin{align}
x[n] &= \frac{1-\alpha^n}{1-\alpha}(1-\alpha)u \\
 &= (1-\alpha^n)u \\
 &= -\alpha^nu + u. \\
\end{align}

With this update equation, we can find the time step, $k_t$, at which we cross threshold, $x_t$. First, we assume the update equation is continuous, substitute in $x_t$ and solve for $t$.

\begin{align}
x_t &= -\alpha^tu + u \\
u - x_t &= \alpha^tu \\
1-\frac{x_t}{u} &= \alpha^t \\
\log_\alpha\left(1-\frac{x_t}{u}\right) &= t \\
\frac{1}{\ln\alpha}\ln\left(1-\frac{x_t}{u}\right) &= t \\
\frac{1}{-\Delta t/\tau}\ln\left(1-\frac{x_t}{u}\right) &= t \\
-\frac{\tau}{\Delta t}\ln\left(1-\frac{x_t}{u}\right) &= t \\
\end{align}

The time step at which we pass threshold will be $k_t=\lceil t\rceil$, so

$$
k_t=\left\lceil-\frac{\tau}{\Delta t}\ln\left(1-\frac{x_t}{u}\right)\right\rceil
$$

and 

$$
f = \frac{1}{\left(k_{ref}+\left\lceil-\frac{\tau}{\Delta t}\ln\left(1-\frac{x_t}{u}\right)\right\rceil\right)\Delta t}.
$$

## Digital LIF neuron with threshold subtraction reset refractory mechanism

Finding the firing rate for the digital LIF neuron with the threshold subtraction reset refractory mechanism is hard and may be impossible. I haven't found it so far. Trying the same approach as the 0 reset refractory mechanism implementation, 

In this case, $x[0]=x_0$ instead of $0$.

\begin{align}
x[0] &= x_0 \\
x[1] &= \alpha x_0 + (1-\alpha)u \\
x[2] &= \alpha^2 x_0 + \alpha(1-\alpha)u + (1-\alpha)u \\
x[3] &= \alpha^3 x_0 + \alpha^2(1-\alpha)u + \alpha(1-\alpha)u + (1-\alpha)u \\
&\vdots \\
x[n] &= \alpha^n x_0 + \sum_{i=0}^{n-1}\alpha^i(1-\alpha)u, \\
\end{align}

so

\begin{align}
x[n] &= \alpha^n x_0 + \frac{1-\alpha^n}{1-\alpha}(1-\alpha)u \\
 &= \alpha^n x_0 + (1-\alpha^n)u \\
 &= \alpha^n(x_0-u) + u. \\
\end{align}

Again, with this update equation, we can find the time step, $k_t$, at which we cross threshold, $x_t$. First, we assume the update equation is continuous, substitute in $x_t$ and solve for $t$.

\begin{align}
x_t &= \alpha^t(x_0 - u) + u \\
x_t - u &= \alpha^t(x_0 - u) \\
\frac{x_t - u}{x_0 - u} &= \alpha^t \\
\log_\alpha\left(\frac{x_t - u}{x_0 - u}\right) &= t \\
-\frac{\tau}{\Delta t}\ln\left(\frac{x_t - u}{x_0 - u}\right) &= t \\
\end{align}

The time step at which we pass threshold will be $k_t=\lceil t\rceil$, so

$$
k_t=\left\lceil-\frac{\tau}{\Delta t}\ln\left(\frac{x_t - u}{x_0 - u}\right)\right\rceil.
$$

Here, we see that $k_t$ depends on $x_0$, and $x_0$ for the next integration period will be the amount that the neuron overshot threshold. 

\begin{align}
\text{overshoot} &= x[k_t]-x_t \\
 &= \alpha^{\left\lceil-\frac{\tau}{\Delta t}\ln\left(\frac{x_t - u}{x_0 - u}\right)\right\rceil}(x_0-u)+u - x_t. \\
\end{align}

If the next $x0$ is different from the current $x0$, the $k_t$ for the next integration period is liable to change. 

This means we cannot simply substitute our above expression for $k_t$ into 

$$
f = \frac{1}{(k_{ref}+k_t)\Delta t}
$$

because $k_t$ can change over time.

### Steady state analysis of threshold subtraction

At each spike the overshoot can change, but we can try to find the steady state conditions. At steady state, the overshoot will match the initial condition of the neuron. That is,

\begin{align}
x[k_t] - x_t &= x_0 \\
\alpha^{k_t}(x_0-u) + u - x_t &= x_0 \\
\alpha^{k_t}(x_0-u) &= x_0 - u  + x_t\\
\alpha^{k_t} &= \frac{x_0 - u + x_t}{x_0-u} \\
\alpha^{k_t} &= 1+\frac{x_t}{x_0-u} \\
k_t &= \log_\alpha\left(1+\frac{x_t}{x_0-u}\right) \\
k_t &= -\frac{\tau}{\Delta t}\ln\left(1+\frac{x_t}{x_0-u}\right) \\
\end{align}

The steady state firing rate will be

\begin{align}
f &= \frac{1}{(k_{ref}+k_t)\Delta t} \\
 &= \frac{1}{k_{ref}\Delta t-\tau\ln\left(1+\frac{x_t}{x_0-u}\right)}.
\end{align}

**Be wary and note that this equation only applies when $k_t$ is an integer.**

If say, I know my desired spike rate in terms of $k_{ref}$ and $k_t$, I can determine $x0-u$:

\begin{align}
k_t &= -\frac{\tau}{\Delta t}\ln\left(1+\frac{x_t}{x_0-u}\right) \\
-k_t\frac{\Delta t}{\tau} &= \ln\left(1+\frac{x_t}{x_0-u}\right) \\
e^{-k_t\Delta t/\tau} &= 1+\frac{x_t}{x_0-u} \\
e^{-k_t\Delta t/\tau}-1 &= \frac{x_t}{x_0-u} \\
x_0-u &= \frac{x_t}{e^{-k_t\Delta t/\tau}-1}. \\
\end{align}

What this says is that given a firing rate $f$, the combination of $x0$ and $u$ that produce $f$ is not unique. It also says that if we fix $f$, $u$, and $x0$ to certain values, the firing rate will be constant. To determine whether we will actually reach the fixed point conditions, we need to look at phase plots.

# Simulating the digital LIF neuron

I make the following points with simulation:
 - Given an constant input $u$, the overshoot will sometimes oscillate and sometimes reach a steady state value.
 - We can predict whether the overshoot will oscillate or reach a stable point from the phase plot.
 - The combination of $x0$ and $u$ that produce a firing rate $f$ is not unique.
 - The digital neuron with the threshold subtraction refractory reset approximates the LIF neuron better than the digital neuron with the 0 refractory reset.
 - We have no equation to theoretically predict the tuning curve of the digital neuron with the threshold subtraction reset whereas we can theoretically predict the tuning curve of the digital neuron with the 0 reset exactly.

```python
%pylab inline
import numpy as np
```

```python
# define the simulation model 
class DigitalNeuron(object):
    def __init__(self, dt=.001, tau=.01, kref=1, xt=1., x0=0., ref_reset='sub'):
        assert ref_reset in ['sub', '0'] # refractory reset mechanism
        self.dt = dt
        self.tau = tau
        self.kref = kref
        self.xt = xt
        self.x0 = x0
        self.ref_reset = ref_reset
        self.x = x0
        self.ref = False
        self.ref_ctr = 0
        self.a = np.exp(-dt/tau)
        
    def reset(self):
        self.x = self.x0
        self.ref = False
        self.ref_ctr = 0
        
    def step(self, u):
        assert isinstance(u, (float, int))
        if self.ref:
            self.ref_ctr += 1
            if self.ref_ctr == self.kref:
                self.ref_ctr = 0
                self.ref = False
        else:
            self.x = self.a*self.x + (1-self.a)*u
            if self.x >= self.xt:
                if self.ref_reset == 'sub':
                    self.x -= self.xt
                elif self.ref_reset == '0':
                    self.x = 0.
                if self.kref > 0:
                    self.ref = True
                return 1
        return 0
    
def continuous_lif_fi(u, tau=.01, tref=.001, xt=1., x0=0.):
    f = np.zeros(u.shape)
    idx = u>xt
    f[idx] = (tref-tau*np.log((u[idx]-xt)/(u[idx]-x0)))**(-1.)
    return f

def digital_0_lif_fi(u, dt=.001, tau=.01, kref=1, xt=1.):
    f = np.zeros(u.shape)
    idx = u>xt
    kt = np.ceil(np.around(-tau/dt*log(1-xt/u[idx]), decimals=10))
    f[idx] = 1./dt*1./(kref + kt) # np.around mitigates floating point error 
    return f
    
def ss_diff(k, dt=.001, tau=.01, xt=1.):
    """returns x0-u"""
    return xt/(np.exp(-k*dt/tau)-1)
```

```python
# phase plot functions
def x0_phase(x0=0., u=2., dt=.001, tau=.01, xt=1.):
    a = np.exp(-dt/tau)
    p = np.ceil(np.around(-tau/dt*log((xt-u)/(x0-u)), decimals=10))
    x0_next = a**p*(x0-u)+u-xt
    return x0_next

from scipy.optimize import brentq
def find_x0_root(u=2., dt=.001, tau=.01, xt=1.):
    f = lambda x0, u, dt, tau, xt: x0_phase(x0=x0, u=u, dt=dt, tau=tau, xt=xt)-x0
    x0_root = brentq(f, 0., xt, args=(u, dt, tau, xt))
    return x0_root
```

```python
# The overshoot can oscillate
dt = .001
nsteps = 100
u=7.34
x = np.zeros(nsteps+1)
t = np.arange(nsteps+1)*dt
neuron = DigitalNeuron(dt)
x0s = [x[0]]
for k in xrange(nsteps):
    spk = neuron.step(u)
    x[k+1]=neuron.x
    if spk:
        x0s.append(x[k+1])
    
figure(figsize=(8,6))
subplot(211)
plot(t, x, '-o')
xlabel('time (s)', fontsize=18)
ylabel('neuron state', fontsize=18)
title('overshoot oscillates over time', fontsize=18)
subplot(212)
plot(x0s, '-o')
xlabel('spike index', fontsize=18)
ylabel('overshoot', fontsize=18);
```

```python
# The overshoot can reach a steady state
dt = .001
nsteps = 100
u=3.11
x = np.zeros(nsteps+1)
t = np.arange(nsteps+1)*dt
neuron = DigitalNeuron(dt)
x0s = [x[0]]
for k in xrange(nsteps):
    spk = neuron.step(u)
    x[k+1]=neuron.x
    if spk:
        x0s.append(x[k+1])
    
figure(figsize=(8,6))
subplot(211)
plot(t, x, '-o')
xlabel('time (s)', fontsize=18)
ylabel('neuron state', fontsize=18)
title('overshoot reaches a steady state', fontsize=18)
subplot(212)
plot(x0s, '-o')
xlabel('spike index', fontsize=18)
ylabel('overshoot', fontsize=18);
```

```python
# we can predict whether x0 will be stable or oscillate
fig = figure(figsize=(12,12))

dt = .001
tau = .01
xt = 1.

a = np.exp(-dt/tau)
umax = xt/(1-a)
u_vals = np.linspace(1., umax, 10)[1:]

x0 = np.linspace(0., xt, 41)[:-1]
for u_idx, u in enumerate(u_vals):
    x0_next = x0_phase(u=u, dt=dt, tau=tau, x0=x0, xt=xt)
    ax = fig.add_subplot(3,3,u_idx+1)
    ax.plot(x0, x0_next, '-o', label='u=%.2f' % u)
    ax.set_xlim(0, xt)
    ax.set_ylim(0, xt)
    ax.plot([0,xt], [0,xt], 'k')
    x0_root = find_x0_root(u=u, dt=dt, tau=tau, xt=xt)
    ax.plot(x0_root, x0_root, 'ro', label='x0=%.2f fixed point' % x0_root)
    ax.legend(loc='upper left')
subplot(338)
xlabel('initial x0', fontsize=18)
subplot(334)
ylabel('next x0', fontsize=18);
subplot(332)
title('x0 phase portraits show x0 fixed point and predict whether x0 will oscillate or settle to the fixed point.'+
      '\nIf the stable point falls in a discontinuous jump in the phase plot, x0 will likely oscillate', 
      fontsize=18);
```

```python
# we can predict the stable x0 given an input u
fig = figure(figsize=(8,6))

dt = .001
tau = .01
xt = 1.

a = np.exp(-dt/tau)
umax = xt/(1-a)
u = np.logspace(log10(xt), log10(umax), 50)[1:]
x0 = np.linspace(0., xt, 20)[:-1]

uv, x0v = np.meshgrid(u, x0)

x0_next = x0_phase(u=uv, dt=dt, tau=tau, x0=x0v, xt=xt)
ax = fig.add_subplot(111)
ax.quiver(uv, x0v, np.zeros(uv.shape), x0_next-x0v)
ax.set_xlabel('u', fontsize=18)
ax.set_ylabel('initial x0', fontsize=18);

x0_roots = []
for u_val in u:
    x0_roots.append(find_x0_root(u=u_val, dt=dt, tau=tau, xt=xt))
ax.plot(u, x0_roots, 'r-o', label='x0 fixed point')
ax.legend(loc='upper right')
ax.set_title('The x0 vector field shows the fixed point of x0 as a function of u', fontsize=18);
```

```python
# u and x0 are not unique for fixed firing rate

fig = figure(figsize=(8,6))
ax = fig.add_subplot(111)

dt = .001
kt = 5
nsteps = 8*kt
x0_u_ss = ss_diff(kt)

u = 2.6
x0 = x0_u_ss + u
x = np.zeros(nsteps+1)
x[0] = x0
t = np.arange(nsteps+1)*dt
neuron = DigitalNeuron(dt, x0=x0)
for k in xrange(nsteps):
    spk = neuron.step(u)
    x[k+1]=neuron.x
ax.plot(t, x, '-o', label='u=%.1f\nx0=%.2f' % (u, x0))
ax.axhline(y=x0, color='k')

u = 2.7
x0 = x0_u_ss + u
x = np.zeros(nsteps+1)
x[0] = x0
t = np.arange(nsteps+1)*dt
neuron = DigitalNeuron(dt, x0=x0)
for k in xrange(nsteps):
    spk = neuron.step(u)
    x[k+1]=neuron.x
ax.plot(t, x, '-o', label='u=%.1f\nx0=%.2f' % (u, x0))
ax.axhline(y=x0, color='k')

ax.axhline(y=1., color='r')
ax.set_xlabel('time (s)', fontsize=18)
ax.set_ylabel('x', fontsize=18)
ax.set_title('x0 and u are not unique for a fixed firing rate', fontsize=18)
ax.legend(loc='right');
```

```python
def fi_experiment(u, dt=.001, tau=.01, xt=1., kref=1, ref_reset='sub'):
    """empirically measure the fi curve"""
    rates = []
    idx = u>xt
    for u_val in u[idx]:
        spk_times = []
        neuron = DigitalNeuron(dt=dt, tau=tau, xt=xt, kref=kref, ref_reset=ref_reset)
        k=0
        while len(spk_times) < 100:
            spk = neuron.step(u_val)
            if spk:
                spk_times.append(k)
            k += 1
        rate = 1./np.mean(np.diff(spk_times[-50:]))
        rates.append(rate)
    f = np.zeros(u.shape)
    f[idx] = np.array(rates)/dt
    return f
```

```python
# fi curve for threshold subtraction refractory mechanism
fig = figure(figsize=(16,6))

dt = .001
tau = .01
kref=1
xt = 1.

a = np.exp(-dt/tau)
umax = xt/(1-a)
u = np.logspace(0, np.log10(umax), 100)
f = fi_experiment(u=u, dt=dt, tau=tau, xt=xt, kref=kref)
f_th = continuous_lif_fi(u, tau=tau, tref=dt*kref, xt=xt)

ax = fig.add_subplot(121)
ax.plot(u, f, '-o', label = 'digital lif w/ subtraction reset')
ax.plot(u, f_th, 'r', label = 'theoretical lif')
ax.legend(loc='upper left')
ax.set_ylabel('f', fontsize=18)
ax.set_xlabel('u', fontsize=18)
ax = fig.add_subplot(122)
ax.plot(u[f>0], 1./(dt*f[f>0]), '-o')
ax.plot(u[f_th>0], 1./(dt*f_th[f_th>0]), 'r')
ax.set_ylabel('period (dt)', fontsize=18)
ax.set_xlabel('u', fontsize=18)
fig.suptitle('Empirically measured fi curve with threshold subtraction refractory reset approximates LIF fi curve'
             , fontsize=18);
```

```python
#fi curve for set 0 refractory mechanism
fig = figure(figsize=(16,6))

dt = .001
tau = .01
kref=1
xt = 1.

a = np.exp(-dt/tau)
umax = xt/(1-a)
u = np.logspace(0, np.log10(umax), 100)
f = fi_experiment(u=u, dt=dt, tau=tau, xt=xt, kref=kref, ref_reset='0')
d_0_lif_th = digital_0_lif_fi(u, dt=dt, tau=tau, kref=kref, xt=xt)
lif_th = continuous_lif_fi(u, tau=tau, tref=dt*kref, xt=xt)

ax = fig.add_subplot(121)
ax.plot(u, f, '-o',  label = 'digital lif w/ 0 reset')
ax.plot(u, lif_th, 'r', label = 'theoretical lif')
ax.plot(u, d_0_lif_th, 'g', label = 'theoretical digital lif w/ 0 reset')
ax.legend(loc='upper left')
ax.set_ylabel('f', fontsize=18)
ax.set_xlabel('u', fontsize=18)
ax = fig.add_subplot(122)
ax.plot(u[f>0], 1./(dt*f[f>0]), '-o')
ax.plot(u[lif_th>0], 1./(dt*lif_th[lif_th>0]), 'r')
ax.plot(u[d_0_lif_th>0], 1./(dt*d_0_lif_th[d_0_lif_th>0]), 'g')
ax.set_ylabel('period (dt)', fontsize=18)
ax.set_xlabel('u', fontsize=18)
fig.suptitle('Empirically measured fi curve with 0 reset refractory mechanism quantizes the LIF neuron fi curve' +
             ' but matches theory.'
             , fontsize=18);
```
