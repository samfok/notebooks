# Notebook description

This notebook looks at the adaptive LIF neuron. The adaptive LIF neuron works like the LIF neuron except that when the neuron spikes, it sends a pulse of negative feedback to itself. Adaptation useful for speeding up the response of a target synapse and reducing the amount of spikes that a neuron sends when the stimulus isn't changing. 

Specifically this notebook looks to answer,
 - What is the steady state firing rate of the adaptive neuron?
 - How does adaptation change the speed of a target synapse's response?

# Summary of results

There is proveably no general, analytical solution for the steady state firing rate of an adaptive neuron. We can approximate the firing rate by setting the feedback to its mean value based on the firing rate of the neuron. This approximation does well at high firing rates but poorly near threshold. Although there is no analytical expression for the firing rate, we also solve for the firing rate numerically without running a simulation. # Background

An adaptive LIF neuron is a modified version of an LIF neuron ([here](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/LIF_neuron.ipynb) is an LIF neuron analysis).

The dynamics of an adaptive LIF neuron are similar to a nonadaptive LIF neuron and are governed by

$$\tau\dot{x}=-x+u_{in}-u_f$$

where 
 - $x$ is the neuron state
 - $\tau$ is the membrane time constant and
 - $u_{in}$ is the input to the neuron
 - $u_f$ is the feedback

when the neuron is not in its refractory period. Unlike the LIF neuron, the adaptive neuron includes a feedback term.

$u_{in}$ can either be controlled directly as an input current or derived from the state of an attached synapse. If $u_{in}$ is derived from a synapse, $u_{in}=\alpha_{syn} x_{syn}$, where $x_{syn}$ is the synaptic state and $\alpha_{syn}$ scales the synaptic state. 

$u_f$ is modeled by an inhibitory synapse that receives a spike whenever the neuron spikes. Therefore, $u_{f}=\alpha_f x_f$ where $x_f$ is the feedback synaptic state and $\alpha_{f}$ scales the feedback synaptic state.

Expanding $u_{in}$ (as a synapse) and $u_f$,

$$\tau\dot{x}=-x+\alpha_{syn}x_{syn}-\alpha_f x_f$$



As with the LIF neuron, the spiking mechanism is triggered by threshold. That is,

$$
x\ge x_t \rightarrow 
  \begin{cases}
    \text{soma outputs a spike} \\
    x :=0 & t_{spk}\le t < t_{spk}+t_{ref} \\
    x_{f}\ := x_{f}+\delta(t_{spk})
  \end{cases}
$$

where
 - $x_t$ is the threshold
 - $t_{spk}$ is the time at which $x(t)$ crosses threshold and
 - $t_{ref}$ is the refractory period
 
In words, when $x$ crosses threshold $x_{t}$, we declare the neuron as having spiked, reset and hold $x$ to $0$ for a refactory period $t_{ref}$, and feed an inhibitory spike back to the adaptive feedback synapse.

# Adaptation speeds up a target synapse's response

We know from the synapse demo [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/synapse/synapse_demo.ipynb) that the time course of a synapse is wholly determined by its time constant and the input firing rate. The expected synapse value will approach the input firing rate exponentially at a rate set by its time constant. The shorter the synaptic time constant, the faster the expected value of the synapse will reach its steady state value.

In addition, the synaptic noise is determined by the input firing rate, the input spike time distribution, and the synaptic time constant. The shorter the synaptic time constant, the noisier the synapse.

Here we see the tradeoff of speed vs accuracy in the synaptic time constant.
 - Shorter synaptic time constants make for faster, noisier synapses.
 - Longer synaptic time constants make for slower, cleaner, synapses.

Are we doomed to live with this tradeoff? No - one thing we didn't mention is changing the input firing rate over time. Imagine that we intend on driving the synapse to a target value. Now, instead of just giving the synapse the target value as input, we initially drive the synapse with a stronger input and then back the input off as the synapse approaches the target value.

```python
# compare saturating exponential
n = 1000
t, dt = np.linspace(0, 8, n, retstep=True)
x1 = 1.*(1-np.exp(-t))
x2 = 2.*(1-np.exp(-t))

u = np.ones(n)
u[t<1.2] = np.linspace(2, 1, len(t[t<1.2]))
decay = np.exp(-dt)
increment = 1-decay
x = np.zeros(n)
for i in xrange(1, n):
    x[i] = decay*x[i-1]+increment*u[i]

fig, ax = plot_continuous(t, x2, plotp={'linewidth':2, 'color':'r', 'label':r'input $=2$'}, figp={'figsize':(8,4)})
fig, ax = plot_continuous(t, x1, plotp={'linewidth':2, 'color':'b', 'label':r'input $=1$'}, ax=ax)
fig, ax = plot_continuous(t, x, plotp={'linewidth':2, 'color':'m', 'label':r'input $=2\rightarrow1$'}, ax=ax, ylim=(0,2.01),
                          axhline=1, axhlinep={'ls':':'},
                          legendp={'loc':'lower right'},
                          xlabel=r'time ($\tau$)', xlabelp={'fontsize':20},
                          ylabel='synapse value', ylabelp={'fontsize':16},
                          title='A stronger input pushes the synapse faster', titlep={'fontsize':16})
```

The red (blue) line is driven by an input of 2 (1). The red line reaches the blue line's steady state value faster than the blue line. The magenta line shows what could happen if we switch from the red line's input to the blue line's input over time. The magenta line reaches the blue line's steady state faster than the blue line and has the same steady state value. Analogously, adaptive neurons work by firing quickly initially and then slowing down to a steady state firing rate. A synapse receiving input from an adaptive neuron would thus reach steady state faster than it would if receiving input from a nonadaptive neuron.

```python
# demo adaptive neuron
T = .2
dt = .0001
tau_m = .01
tref = .005
xt = 1.
af = 5e-2
tau_f = 1e-1

ytop = xt*1.2  # top of plot

# timeseries for a constant input
u_vals = [1.3, 2.5]
fig = None
subplots = [(2, 2, 1), (2, 2, 3), (2, 2, 2), (2, 2, 4)]
for idx, u_val in enumerate(u_vals):
    nsteps = int(np.ceil(T/dt))
    t = np.arange(nsteps)*dt
    u = np.zeros(nsteps)+u_val
    lif_spk_times, lif_x = run_lifsoma(dt, u, tau_m, tref, xt, ret_state=True)
    alif_spk_times, alif_x, alif_xf = run_alifsoma(dt, u, tau_m, tref, xt, af=af, tauf=tau_f,
                                                   ret_state=True, ret_fstate=True)

    if fig is None:
        fig, ax = plot_continuous(t, lif_x, plotp={'c':'k', 'lw':1, 'ls':':', 'label':'nonadaptive LIF state'},
                                  figp={'figsize':(16,10)}, subplotp=subplots[2*idx])
    else:
        fig, ax = plot_continuous(t, lif_x, plotp={'c':'k', 'lw':1, 'ls':':', 'label':'nonadaptive LIF state'}, fig=fig, 
                                  subplotp=subplots[2*idx])

    plot_continuous(t, alif_x, plotp={'color':'b','linewidth':2, 'label':'adaptive LIF state'}, ax=ax,
                    axhline=xt, axhlinep={'color':'r', 'ls':':'},
                    legendp={'loc':'lower right'},
                    ylabel=r'$x$', ylabelp={'fontsize':24},
                    title=r'$u_{in}=%.2f$' % u_val, titlep={'fontsize':20})
    ax.set_ylim(ax.get_ylim()[0], ytop)

    fig, ax = plot_continuous(t, alif_xf, plotp={'color':'b'}, fig=fig, subplotp=subplots[2*idx+1],
                              xlabel=r'$t$ (s)', xlabelp={'fontsize':20},
                              ylabel=r'$x_f$', ylabelp={'fontsize':24},)
fig.suptitle('Adaptation slows down the initial firing rate', fontsize=20);
```

In the upper plots, the black (blue) traces show the state of nonadaptive (adaptive) LIF neurons in response to a constant input. The lower plots show the adaptive feedback as the adaptive neuron begins to spike.

We can see that the interspike intervals increase as adaptation ($x_f$) kicks in and that the adaptive feedback behaves like like a synapse. In fact, if the adaptive neuron projects to a synapse (that has no other inputs) with the same time constant as the adaptation, the synapse will exactly match the adative feedback.

This behavior suggests a strategy for estimating when the adaptive neuron has reached steady state. First, let's pretend we knew the adaptive neuron's steady state firing rate (which is the next [section](#Steady-state-firing-rate) anyways). We know from the synapse demo [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/synapse/synapse_demo.ipynb) that a synapse's expected state approaches the input spike rate exponentially with the synaptic time constant. As adaptation behaves identically to a synapse, we know that the expected adaptation state approaches the neuron's firing rate exponentially with the adaptive feedback time constant. We also know that adaptation decreases the neuron's firing rate, so the neuron initially fires faster than at steady state. Therefore, the exepcted adaptive state approaches the steady state firing rate faster than the expected state of a hypothetical synapse receiving spikes at the steady state rate. This means that we need observe the synapse for a length of time determined by the adaptive feedback time constant. More concretely, running the adaptive neuron for something like $5\tau_f$ should suffice for declaring that the adaptation, and hence the neuron, has reached steady state. Since the firing rate is uniform at steady state, we need only measure a single interspike interval to determine the spike rate.

# Steady state firing rate

As with the nonadaptive LIF neuron, we try to find $t_{spk}$ in

$$f=\frac{1}{t_{ref}+t_{spk}}$$

We will see that there is no general, analytical solution to the steady state firing rate of the adaptive neuron. As stated above, the neuron dynamics (outside of the refractory period) are governed by

\begin{align}
\tau\dot{x} &= -x+u \\
\dot{x} + \frac{1}{\tau}x &= \frac{1}{\tau}u \\
\end{align}

In contrast to the LIF case, $u$ is now time varying because of adaptation. That is, $u(t)=u_{in}-u_f(t)$, where $u_f(t)=ke^{-t/\tau_f}$ and $k$ is the amplitude of the feedback scaled by $\alpha_f$ after the refractory period ends. Note that $k$ is not really a free parameter but rather depends on the firing rate of the neuron. That is, $k=k(f, \alpha_f, t_{ref})$.

```python
# show what k is
T = .4
dt = .0001
tau_m = .01
tref = .005
xt = 1.
af = 5e-2
tau_f = 1e-1

u_val = 1.3
nsteps = int(np.ceil(T/dt))
t = np.arange(nsteps)*dt
u = np.zeros(nsteps)+u_val
alif_spk_times, alif_x, alif_xf = run_alifsoma(dt, u, tau_m, tref, xt, af=af, tauf=tau_f,
                                               ret_state=True, ret_fstate=True)
isi = alif_spk_times[-1]-alif_spk_times[-2]
f = 1./isi
k = af*np.exp(-tref/tau_f)/(tau_f*(1-np.exp(-1./f/tau_f)))
fig, ax = plot_continuous(t, alif_x, plotp={'color':'b','linewidth':2}, subplotp=(2,1,1), figp={'figsize':(12,6)},
                          axhline=xt, axhlinep={'color':'r', 'ls':':', 'label':r'$x_t$'},
                          axvline=alif_spk_times[2]+tref, axvlinep={'c':'k'},
                          legendp={'loc':'lower right', 'fontsize':18},
                          ylabel=r'$x$', ylabelp={'fontsize':24},
                          title=r"$k$ is the steady state value of feedback $(u_f)$ at the refractory period's end",
                          titlep={'fontsize':18})
ax.set_ylim(ax.get_ylim()[0], 1.2)
fig, ax = plot_continuous(t, af*alif_xf, plotp={'color':'b'}, fig=fig, subplotp=(2,1,2),
                          axhline=k, axhlinep={'color':'r', 'label':r'$k$'},
                          axvline=alif_spk_times[2]+tref, axvlinep={'c':'k'},
                          legendp={'loc':'lower right', 'fontsize':18},
                          xlabel=r'$t$ (s)', xlabelp={'fontsize':20},
                          ylabel=r'$u_f$', ylabelp={'fontsize':24},)
```

See that $k$ intersects the feedback $u_f$ at the end of the neuron's refractory period denoted by the black line. The neuron's firing rate will be uniform and the feedback is equivalent to a synapse that receives spikes whenever the neuron spikes, so we can express $k$ in terms of $f$, $\alpha_f$, and $t_{ref}$.

\begin{align}
k &= k(f, \alpha_f, t_{ref}) \\
 &= \frac{\alpha_f e^{-t_{ref}/\tau_f}}{\tau_f(1-e^{-1/f\tau_f})} \\
k &= \frac{\alpha_f e^{-t_{ref}/\tau_f}}{\tau_f(1-e^{-(t_{ref}+t_{spk})/\tau_f})} \\
\end{align}

where
 - $\alpha_f$ is the user-controlled the feedback strength,
 - $e^{-t_{ref}/\tau_f}$ is the decay of the feedback during the refractory period, and
 - $\left(\tau_f(1-e^{-1/f\tau_f})\right)^{-1}$ is the maximum amplitude of a synapse receiving uniform input spikes (derived in the uniform synapse [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/synapse/synapse_SNR_uniform.ipynb#Summary-of-findings)).

Having analyzed $k$, we set it aside for now and assume that $k$ is a free parameter in our firing rate derivation.

Set integrating factor 

\begin{align}
\gamma(t) &= e^{\int \frac{1}{\tau}dt} \\
 &= e^{t/\tau} \\
\end{align}

where the integrating constant will be folded into the overall solution. 

\begin{align}
\dot{x} + \frac{1}{\tau}x &= \frac{1}{\tau}u \\
\frac{d}{dt}x + \frac{1}{\tau}x &= \frac{1}{\tau}u \\
\gamma\frac{d}{dt}x + \frac{1}{\tau}\gamma x &= \frac{1}{\tau}\gamma u \\
\gamma\frac{d}{dt}x + \left(\frac{d}{dt}\gamma\right) x &= \frac{1}{\tau}\gamma u \\
(\gamma x)' &= \frac{1}{\tau}\gamma u \\
\gamma(t) x(t) &= \int\frac{1}{\tau}\gamma u(t)dt + c \\
e^{t/\tau} x(t) &= \int\frac{1}{\tau}e^{t/\tau}u(t)dt + c \\
 &= \int\frac{1}{\tau}e^{t/\tau}(u_{in}-u_f(t))dt + c \\
 &= \frac{u_{in}}{\tau}\int e^{t/\tau}dt - \frac{1}{\tau}\int e^{t/\tau}u_f(t)dt + c \\
 &= \frac{u_{in}}{\tau}\int e^{t/\tau}dt - \frac{k}{\tau}\int e^{t/\tau}e^{-t/\tau_f}dt + c \\
 &= \frac{u_{in}}{\tau}\int e^{t/\tau}dt - \frac{k}{\tau}\int e^{t(1/\tau-1/\tau_f)}dt + c \\
e^{t/\tau}x(t) &= u_{in}e^{t/\tau} - \frac{k}{\tau(1/\tau-1/\tau_f)}e^{t(1/\tau-1/\tau_f)} + c \\
 &= u_{in}e^{t/\tau} - \frac{k\tau_f}{\tau_f-\tau}e^{t(1/\tau-1/\tau_f)} + c \\
x(t) &= u_{in} + \frac{k\tau_f}{\tau-\tau_f}e^{t(1/\tau-1/\tau_f)}e^{-t/\tau} + ce^{-t/\tau} \\
 &=  u_{in} + \frac{k\tau_f}{\tau-\tau_f}e^{-t/\tau_f} + ce^{-t/\tau} \\
\end{align}

Using the initial condition of $x(0)=0$,

\begin{align}
0 &= u_{in}+\frac{k\tau_f}{\tau-\tau_f} + c \\
c &= -\frac{k\tau_f}{\tau-\tau_f} - u_{in} \\
\end{align}

so

\begin{align}
x(t) &= u_{in}+\frac{k\tau_f}{\tau-\tau_f}e^{-t/\tau_f}+\left(-\frac{k\tau_f}{\tau-\tau_f}-u_{in}\right)e^{-t/\tau} \\
x(t) &= \overbrace{u_{in}(1-e^{-t/\tau})}^\text{from the input} + 
    \overbrace{\frac{k\tau_f}{\tau-\tau_f}\left(e^{-t/\tau_f}-e^{-t/\tau}\right)}^\text{from the adaptation} \\
\end{align}

Comparing this with the nonadaptive neuron, we see the familiar term relating the dynamics to the constant input. We now also see the dynamics due to adaptation. At this point, we would try to find $t_{spk}$ such that $x(t_{spk})=x_t$. 

$$x_t = u_{in}(1-e^{-t_{spk}/\tau})+\frac{k\tau_f}{\tau-\tau_f}\left(e^{-t_{spk}/\tau_f}-e^{-t_{spk}/\tau}\right)$$

However, there is no general way to algebraically solve this equation for $t_{spk}$. To show this, we first express $\tau_f$ as a scaling of $\tau$ with $\tau_f = \tau/\beta$ for some $\beta>0$. Substituting,

\begin{align}
x_t &= u_{in}(1-e^{-t_{spk}/\tau})+\frac{k\tau/\beta}{\tau-\tau/\beta}
    \left(e^{-\beta t_{spk}/\tau}-e^{-t_{spk}/\tau}\right) \\
 &= u_{in}(1-e^{-t_{spk}/\tau})+\frac{k}{\beta-1}\left(e^{-\beta t_{spk}/\tau}-e^{-t_{spk}/\tau}\right) \\
\end{align}

and letting $y=e^{-t_{spk}/\tau}$,

\begin{align}
x_t &= u_{in}(1-y)+\frac{k}{\beta-1}(y^\beta-y) \\
 &= u_{in}-u_{in}y+\frac{k}{\beta-1}y^\beta-\frac{k}{\beta-1}y \\
0 &= \frac{k}{\beta-1}y^\beta - \left(u_{in}+\frac{k}{\beta-1}\right)y + (u_{in}-x_t) \\
\end{align}

which looks like a polynomial of degree $\beta$. If we could somehow find the roots of the equation, then it would be a simple matter of back-substituting for $y$ and solving for $t_{spk}$. However $\beta$ could be any number greater than $0$, so the equation is only contained within the more general class of [exponential polynomials](http://math.stackexchange.com/questions/793109/polynomials-with-non-integer-exponents). Even if we restricted ourselves to integer values of $\beta$ so that the equation was a polynomial, we know from the [Abel-Ruffini theorem](https://en.wikipedia.org/wiki/Abel%E2%80%93Ruffini_theorem) that for $\beta\ge5$, there is no algebraic solution to the polynomial. Note that this is not to say that we could never find a solution. We could still probably find solutions for specific values of $\beta$.

For example, what if we match the adaptive time constant to soma time constant so that $\beta=1$?

\begin{align}
0 &= \frac{k}{1-1}y - \left(u_{in}+\frac{k}{1-1}\right)y + (u_{in}-x_t) \\
 &= \left(\frac{k}{1-1}-\frac{k}{1-1}-u_{in}\right)y + (u_{in}-x_t) \\
\end{align}

The $1-1$ in the denominators look problematic, but we can make progress by returning to the original formuation before substituting $y$. That is,

\begin{align}
x_t &= u_{in}(1-e^{-t_{spk}/\tau})+\frac{k\tau_f}{\tau-\tau_f}\left(e^{-t_{spk}/\tau_f}-e^{-t_{spk}/\tau}\right) \\
 &= u_{in}(1-e^{-t_{spk}/\tau})+\frac{k\tau}{\tau-\tau}\left(e^{-t_{spk}/\tau}-e^{-t_{spk}/\tau}\right) 
 & \tau_f=\tau \\
\end{align}

This again looks problematic because $\tau-\tau=0$ divides $e^{-t_{spk}/tau}-e^{-t_{spk}/tau}=0$. However, both terms go to $0$, so we can use L'Hopital's rule to find their limit.

\begin{align}
\lim_{\tau_f\rightarrow\tau}\frac{k\tau_f}{\tau-\tau_f}\left(e^{-t/\tau_f}-e^{-t/\tau}\right) 
 &= k\lim_{\tau_f\rightarrow\tau}\frac{\tau_f\left(e^{-t/\tau_f}-e^{-t/\tau}\right)}{\tau-\tau_f} \\
 &= k\lim_{\tau_f\rightarrow\tau}\frac{\left(\tau_f\left(e^{-t/\tau_f}-e^{-t/\tau}\right)\right)'}{(\tau-\tau_f)'} 
    &\text{L'Hopital's Rule} \\
 &= k\lim_{\tau_f\rightarrow\tau}\frac{\left(e^{-t/\tau_f}-e^{-t/\tau}\right)+\tau_fe^{-t/\tau_f}\frac{t}{\tau_f^2}}{-1} \\
 &= -\frac{k}{\tau}te^{-t/\tau} \\
\end{align}


so when $\tau_f=\tau$ we have

$$x(t) = u_{in}(1-e^{-t/\tau})-\frac{k}{\tau}te^{-t/\tau}$$

Again, attempting to find $t_{spk}$,

\begin{align}
x_t &= u_{in}(1-e^{-t_{spk}/\tau})-\frac{k}{\tau}t_{spk}e^{-t_{spk}/\tau} \\
\end{align}

we find ourselves stymied in our quest for an analytical solution for $t_{spk}$. Mathematica says the solution is 

$$t_{spk}=-\frac{u_{in}}{k}-\tau W\left(-\frac{e^{-u/(k\tau)}(u_{in}-x_t)}{k\tau}\right)$$

where $W(\cdot)$ is the [product logarithm](http://reference.wolfram.com/language/ref/ProductLog.html) (a.k.a. the [Lambert W function](https://en.wikipedia.org/wiki/Lambert_W_function)), but how it found the solution is beyond me [If the solution is not beyond you, find me!]. At this point, I think it's time we just solve the equation numerically. 

First though, let's check that the theoretical solution,

$$
x(t) = 
 \begin{cases}
    u_{in}(1-e^{-t/\tau})+\frac{k\tau_f}{\tau-\tau_f}\left(e^{-t/\tau_f}-e^{-t/\tau}\right) &\quad \tau\neq\tau_f \\
    u_{in}(1-e^{-t/\tau})-\frac{k}{\tau}te^{-t/\tau} &\quad \tau=\tau_f \\
 \end{cases}
$$

to the adaptive neuron's differential equation matches the simulation.

```python
# check the theory with simulation

# theory
def th_lif_x(t, u, tau):
    return u*(1-np.exp(-t/tau))

def th_alif_x(t, uin, k, tau, tauf):
    if tau != tauf:
        x = uin*(1.-np.exp(-t/tau)) + k*tauf/(tau-tauf)*(np.exp(-t/tauf)-np.exp(-t/tau))
    elif tau == tauf:  # because Python doesn't know LHopital's Rule
        x = uin*(1.-np.exp(-t/tau)) - k/tau*t*np.exp(-t/tau)
    return x

# simulation
dt=.0001
T = .5
N = int(np.ceil(T/dt))
t = np.arange(N)*dt

taum = .1
k = 2.
tauf_ = [.01, .05, .1, .2]

uin = 2.
for tauf in tauf_:
    uf = k*np.exp(-t/tauf)
    uadapt = uin - uf

    decay = np.exp(-dt/taum)
    increment = 1-decay

    x = np.zeros(N)
    xadapt = np.zeros(N)
    for i in xrange(1,N): 
        x[i] = decay*x[i-1]+increment*uin
        xadapt[i] = decay*xadapt[i-1]+increment*uadapt[i]

    xth = th_lif_x(t, uin, taum)
    xadaptth = th_alif_x(t, uin, k, taum, tauf)

    figure(figsize=(14,3.5))
    subplot(121)
    plot(t, uadapt, 'k')
    axhline(uin, c='k', ls=':')
    ylabel(r'$u_{in}-u_f(t)$', fontsize=20)
    xlabel(r'$t$', fontsize=18)
    title(r'$\tau=%.2f$ , $\tau_f=%.2f$' % (taum, tauf), fontsize=20)

    subplot(122)
    plot(t, x, 'b', label='nonadaptive simulation')
    plot(t, xth, 'b:', lw=4, label='nonadaptive theory')
    plot(t, xadapt, 'r', label='adaptive simulation')
    plot(t, xadaptth, 'r:', lw=4, label='adaptive theory')
    legend(loc='lower right')
    ylabel(r'$x(t)$', fontsize=20)
    xlabel(r'$t$', fontsize=18)
```

The differential equation simulation result lines up on top of the theoretical differential equation solution. 

To find the steady state firing rate, recall that

$$f=\frac{1}{t_{ref}+t_{spk}}$$

and that 

$$
x_t = 
\begin{cases}
    u_{in}(1-e^{-t_{spk}/\tau})+\frac{k\tau_f}{\tau-\tau_f}\left(e^{-t_{spk}/\tau_f}-e^{-t_{spk}/\tau}\right)
    &\quad \tau\neq\tau_f \\
    u_{in}(1-e^{-t_{spk}/\tau})-\frac{k}{\tau}t_{spk}e^{-t_{spk}/\tau} &\quad \tau=\tau_f \\
\end{cases}
$$

Our goal is to solve these equations numerically. First, let's remove our assumption that $k$ is a free parameter. Recall that

$$k = \frac{\alpha_f e^{-t_{ref}/\tau_f}}{\tau_f(1-e^{-(t_{ref}+t_{spk})/\tau_f})}$$

Substituting for $k$,

\begin{align}
x_t &= u_{in}(1-e^{-t_{spk}/\tau})+
    \frac{\alpha_f e^{-t_{ref}/\tau_f}\left(e^{-t_{spk}/\tau_f}-e^{-t_{spk}/\tau}\right)}
    {(1-e^{-(t_{ref}+t_{spk})/\tau_f})(\tau-\tau_f)} & \tau\neq\tau_f \\
x_t &= u_{in}(1-e^{-t_{spk}/\tau})-
    \frac{\alpha_f t_{spk}e^{-(t_{ref}+t_{spk})/\tau}}{\tau^2(1-e^{-(t_{ref}+t_{spk})/\tau})} & \tau=\tau_f \\
\end{align}

Our approach uses the fact that $f$ monotonically increases with $u_{in}$. Similarly, $(t_{spk})$ monotonically decreases with $u_{in}$. Rearranging in terms of $u_{in}$,

\begin{align}
u_{in} &= \frac{1}{1-e^{-t_{spk}/\tau}}
    \left(x_t-
    \frac{\alpha_f e^{-t_{ref}/\tau_f}\left(e^{-t_{spk}/\tau_f}-e^{-t_{spk}/\tau}\right)}
    {(1-e^{-(t_{ref}+t_{spk})/\tau_f})(\tau-\tau_f)}\right) & \tau\neq\tau_f \\
u_{in} &= \frac{1}{1-e^{-t_{spk}/\tau}}
    \left(x_t+
    \frac{\alpha_f t_{spk}e^{-(t_{ref}+t_{spk})/\tau}}{\tau^2(1-e^{-(t_{ref}+t_{spk})/\tau})}\right) & \tau=\tau_f \\
\end{align}

We binary search (a.k.a. the [bisection method](http://en.wikipedia.org/wiki/Bisection_method) in root-finding parlance) over the range of possible firing rates to find the firing rate that corresponds to each input $u_{in}$. 

```python
fig = figure(figsize=(16,12))
axs = [fig.add_subplot(2,2,i+1) for i in xrange(4)]
tauf = [.001, .01, .1, 1.]
af = [1e-3, 1e-2, 1e-1, 1e0]
sim_vs_num_tauf(dt=.0001, T=3., max_u=5., taum=.01, tref=.005, xt=1., af=1e-1, tauf=tauf, ax=axs[0])
sim_vs_num_tauf(dt=.0001, T=3., max_u=5., taum=.01, tref=.005, xt=1., af=5e-3, tauf=tauf, ax=axs[1])
sim_vs_num_af(dt=.0001, T=3., max_u=5., taum=.01, tref=.005, xt=1., af=af, tauf=5e-3, ax=axs[2])
sim_vs_num_af(dt=.0001, T=3., max_u=5., taum=.01, tref=.005, xt=1., af=af, tauf=1e-1, ax=axs[3])
```

The firing rates found with simulation (circles) agree with the firing rates found numericaly (blue lines) across a wide range of $\alpha_f$ and $\tau_f$. 

The top plots show firing rates for fixed $\alpha_f$ and varying $\tau_f$. The bottom plots show firing rates for fixed $\tau_f$ and varying $\alpha_f$. Increasing either $\alpha_f$ or $\tau_f$ reduces the firing rates. 

The upper left plot shows that with sufficient $\alpha_f$, increasing $\tau_f$ linearizes the firing rate curves. The upper right plot shows that without sufficient $\alpha_f$, increasing $\tau_f$ does not have the same linearizing effect.

The lower left plot shows that with short $\tau_f$, increasing $\alpha_f$ flattens the high firing rates without linearizing the firing rate curve. The lower right plot shows that with sufficient $\tau_f$, increasing $\alpha_f$ linearizes the tuning curve.

## Steady state firing rate - mean feedback approximation

One way of approximating the adaptive neuron firing rate is to assume that the feedback is constant at its mean value. That is, we can assume $u_f = E[u_f(t)]$ where the expectation is taken over time. The validity of this assumption is discussed [below](#When-mean-value-approximation-is-valid). Assuming constant feedback allows us to lump the feedback with $u_{in}$ and reuse the firing rate equation for a nonadaptive LIF neuron with constant input.

\begin{align}
f &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u}\right)} & u>x_t \\
 &=  \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-u_f}\right)} & u_{in}-u_f>x_t \\
\end{align}

Since $u_f$ is derived from a synapse but only takes effect after the refractory period ends, $E[u_f]=\alpha_f e^{-t_{ref}/\tau_f}E[x_f]=\alpha_f e^{-t_{ref}/\tau_f}f$ (The synapse demo [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/synapse/synapse_demo.ipynb) shows $E[x_f]=f$). Let $\alpha_f'=\alpha_f e^{-t_{ref}/\tau_f}$. With these approximations,

\begin{align}
f &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-u_f}\right)} & \text{assuming $u_{in}$ and $u_f$ are constant} \\
 &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-E[u_f]}\right)} & u_f=E[u_f] \\
 &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-\alpha_f'f}\right)} & E[u_f]=\alpha_f'f \\
\end{align}


While we can't rearrange this equation to solve for $f$ [Try Mathematica], we can linearly approximate the solution or find the solution numerically. First, though, let's visualize the system of equations.

```python
tau_m = .01
tau_f = .005
tref = .005
xt = 1.
af = .02
af *= np.exp(-tref/tau_f)

th_max_f = 1./tref  # theoretical maximum firing rate
max_u = 5.  # maximum u_in to consider
max_f = th_lif_fi(max_u, tau_m, tref, xt)
f = np.linspace(5., max_f, 19)
u = th_lif_if(f, tau_m, tref, xt)  # linear in firing rate

cc = [(float(i)/len(u), 0, 0) for i in xrange(len(u))]
fig = figure(figsize=(16,6))

ax = fig.add_subplot(121)
_u = np.array(sort(np.linspace(0, max_u, 100).tolist() + [xt, 1.001*xt, 1.01*xt, 1.1*xt]))
ax.plot(_u, th_lif_fi(_u, tau_m, tref, xt), 'k', linewidth=2)
ax.set_color_cycle(cc)
for idx, u_val in enumerate(u[u>xt]):
    f = th_lif_fi(u_val, tau_m, tref, xt)
    ax.plot([u_val, u[-1]], [f, f], c=cc[idx])
    ax.plot([u_val, u_val], [0, f], c=cc[idx], ls=':')
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max_u*1.001)
ax.set_xlabel(r'$u_{in}$', fontsize=20)
ax.set_ylabel(r'$f(u_{in})$ (spks / s)', fontsize=20)
ax.set_title('Nonadaptive Neuron ' + r'$(u_f=0)$', fontsize=16)

f = np.linspace(0, th_max_f, 201)
uf = af*f

ax = fig.add_subplot(122, color_cycle=cc)
for idx, u_val in enumerate(u):
    net_u = u_val - uf
    f_ss = th_lif_fi(net_u, tau_m, tref, xt)
    line = ax.plot(uf, f_ss)[0]
    if u_val in (u[0], u[-1]):
        line.set_label(r'$u_{in}=%.2f$' % u_val)
ax.plot(uf, f, 'b', label=r'$u_f=\alpha_f e^{-t_{ref}/\tau_f}f$')
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max(uf))
ax.legend(loc='upper left', fontsize=16)
ax.set_xlabel(r'$u_f$', fontsize=20)
ax.set_ylabel(r'$f(u_f;\ u_{in})$ (spks / s)', fontsize=20)
ax.set_title('Adaptatation reduces the firing rate', fontsize=16);
```

The left plot shows the usual nonadaptive LIF firing rate equation. The right plot shows how adaptive feedback ($u_f$) reduces the firing rate for fixed $u_{in}$. Every point $(u_{in}, f(u_{in}))$ on the left plot corresponds to a point $(0, f(u_{in},0))$ on the right plot's y-axis, which describes the firing rate of a neuron without any adaptation $(u_f=0)$. Further, on the right plot,
 - the firing rate decreases as inhibitory feedback increases, and
 - inhibitory feedback increases linearly with firing rate ($u_f=\alpha_ff$).

The steady state firing rate is the intersection between the $f(u_{in}, u_f)$ curve and the $u_f=\alpha_f f$ line. To see why, consider that 
 - if we decrement $u_f$ from the intersection point (i.e. move left along a fixed $u_{in}$ curve), the firing rate increases, which increases $u_f$, which cancels our initial decrement. Similarly, 
 - if we incrememt $u_f$ from the intersection point (i.e. move right along a fixed $u_{in}$ curve), the firing rate decreases, which decreases $u_f$, which cancels our initial increment.

### Mean feedback approximation - linear approximation

We can use the linear approximation of the tuning curve equation from the LIF [notebook](http://nbviewer.ipython.org/github/fragapanagos/notebooks/blob/master/neuron/LIF_neuron.ipynb#Linear-approximation-of-the-LIF-firing-rate-equation) to estimate the effect of adaptation on the tuning curve.

\begin{align}
f &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{u_{in}-\alpha_f'f}\right)} \\
 &\approx k_0 + k_1(u_{in}-\alpha_f'f) \\
\end{align}

where

\begin{align}
k_0 &= \frac{1}{t_{ref}-\tau\ln\left(1-\frac{x_t}{a}\right)} - k_1a\\
k_1 &= \frac{\tau x_t}{\left(t_{ref}-\tau\ln\left(1-\frac{x_t}{a}\right)\right)^2a(a-x_t)} \\
\end{align}

$a$ is the region of $u_{in}-\alpha_f'f$ that we are approximating around. We will set $a=u_{in}$. That is, we are linearly approximating from the nonadaptive ($u_f=0$) LIF firing rate equation.

\begin{align}
f &= k_0 + k_1(u_{in}-u_f) \\
 &= k_0 + k_1u_{in}-k_1u_f & \text{linearly approximate $f(u_f; u_{in})$}\\
 &= k_0 + k_1u_{in}-k_1\alpha_f'f & u_f=\alpha_f'f\\
(1+k_1\alpha_f')f &= k_0 +k_1u_{in} \\
f &= \frac{k_0 +k_1u_{in}}{1+k_1\alpha_f'} & \text{intersection of $f(u_f; u_{in})$ and $u_f=\alpha_f'f$}\\
\end{align}


```python
tau_m = .01
tau_f = .005
tref = .005
xt = 1.
af = .02

th_max_f = 1./tref  # theoretical maximum firing rate
max_u = 5.  # maximum u_in to consider
max_f = th_lif_fi(max_u, tau_m, tref, xt)
f = np.linspace(5., max_f, 15)
u_in = th_lif_if(f, tau_m, tref, xt)  # linear in firing rate

uf_f = np.linspace(0, th_max_f, 201)
uf = af*np.exp(-tref/tau_f)*uf_f

k0, k1 = taylor1_lif_k0_k1(u_in, tau_m, tref, xt)
taylor1_f_uf = np.zeros((u_in.shape[0], uf.shape[0]))
for idx, u_val in enumerate(u_in):
    taylor1_f_uf[idx, :] = k0[idx]+k1[idx]*u_val-k1[idx]*uf
taylor1_af = taylor1_alif_fi(u_in, tau_m, tref, xt, af, tau_f)

cc = [(float(i)/len(u_in), 0, 0) for i in xrange(len(u_in))]
alpha = .5
fig = figure(figsize=(16,6))

ax = fig.add_subplot(121)
_u = np.array(sort(np.linspace(0, max_u, 100).tolist() + [xt, 1.001*xt, 1.01*xt, 1.1*xt]))
idx = _u>xt
k0, k1 = taylor1_lif_k0_k1(_u[idx], tau_m, tref, xt)
_taylor1_af = taylor1_alif_fi(_u, tau_m, tref, xt, af, tau_f)

ax.plot(_u, th_lif_fi(_u, tau_m, tref, xt), 'k', linewidth=2, label='nonadaptive LIF')
ax.plot(_u, _taylor1_af, 'b', linewidth=2, label='adaptive LIF')
ax.plot(u_in, taylor1_af, 'bo', ms=8, alpha=alpha)
for idx, u_val in enumerate(u_in):
    f = th_lif_fi(u_val, tau_m, tref, xt)
    ax.plot([u_val, u_in[-1]], [f, f], c=cc[idx])
    ax.plot([u_val, u_val], [0, f], c=cc[idx], ls=':')
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max_u*1.001)
ax.legend(loc='upper left')
ax.set_xlabel(r'$u_{in}$', fontsize=20)
ax.set_ylabel(r'$f(u_{in})$ (spks / s)', fontsize=20)

ax = fig.add_subplot(122)
for idx, u_val in enumerate(u_in):
    net_u = u_val - uf
    f_ss = th_lif_fi(net_u, tau_m, tref, xt)
    line = ax.plot(uf, f_ss, color=cc[idx])[0]
    if u_val in (u_in[0], u_in[-1]):
        line.set_label(r'$u_{in}=%.2f$' % u_val)
    ax.plot(uf, taylor1_f_uf[idx, :], color=cc[idx], ls=':')
iter_uf = af*np.exp(-tref/tau_f)*taylor1_af
ax.plot(iter_uf, taylor1_af, 'bo', ms=8, alpha=alpha)
ax.plot(uf, uf_f, 'b', label=r"$u_f=\alpha_f'f$")
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max(uf))
ax.legend(loc='upper left', fontsize=16)
ax.set_ylim(0, th_max_f)
ax.set_xlabel(r'$u_f$', fontsize=20)
ax.set_ylabel(r'$f(u_f;\ u_{in})$ (spks / s)', fontsize=20)
fig.suptitle('Linearly approximation of the adaptive LIF firing rate', fontsize=16);
```

The blue circles on the left plot correspond to the blue circles on the right plot. The dashed lines correspond to our linear approximations of the effect of the mean value approximated feedback. Linear approximation performs poorly at low firing rates but well at high firing rates.

## When mean value approximation is valid

First order, linear differential equations can be expressed as

$$\dot{x} + p(t)x = g(t)$$

Here, we look at when we can replace $g(t)$ with its mean value. That is,

$$\dot{x} + p(t)x = E[g(t)]$$

The mean value of $g$ is defined as

$$
E[g(t)] = \frac{1}{\Delta t}\int_{\Delta t} g(t)dt
$$

over some period $\Delta t$. If $\Delta t$ starts at $0$ and ends at $t$,

$$
E[g(t)] = \frac{1}{t}\int_{0}^t g(t')dt'
$$

so 

$$
tE[g(t)] = \int_{0}^t g(t)dt
$$

Setting integrating factor $\mu(t)=e^{\int p(t)dt}$,

\begin{align}
\dot{x} + px(t) &= g \\
\mu\dot{x} + p\mu x &= \mu g \\
(x \mu)' &= \mu g \\
x(t)\mu(t) &= \int \mu(t)g(t)dt + c \\
x(t) &= \frac{\int \mu(t)g(t)dt + c}{\mu(t)} \\
\end{align}

we can see two cases where setting $g(t)$ to its mean value will produce the same dynamics in $x(t)$ as letting $g(t)$ vary with time. The trivial case is when $g(t)$ is constant. When $g(t)=g$,

$$E[g(t)] = g$$

so

\begin{align}
x(t) &= \frac{\int \mu(t)g(t)dt + c}{\mu(t)} \\
 &= \frac{\int \mu(t)gdt + c}{\mu(t)} \\
 &= \frac{\int \mu(t)E[g(t)]dt + c}{\mu(t)} \\
\end{align}

The other case is when $\mu(t)$ is a constant, which occurs when $p(t)=0$ because

\begin{align}
\mu(t) &= e^{\int p(t)dt} \\
 &= e^{\int 0dt} \\
 &= e^{k} \\
 &= \mu \\
\end{align}

so when $\mu(t)=\mu$,

\begin{align}
x(t) &= \frac{\int \mu(t)g(t)dt + c}{\mu(t)} \\
 &= \frac{\int \mu g(t)dt + c}{\mu} \\
 &= \int g(t)dt + c/\mu \\
 &= \int g(t)dt + c & c:=c/\mu \\
\end{align}

Recognizing that 

$$
tE[g(t)] = \int_{0}^t g(t)dt
$$

We see that we can replace $g(t)$ with $E[g(t)]$ without affecting the dynamics of $x(t)$.

\begin{align}
x(t) &= \int g(t)dt + c \\
 &= \int E[g(t)]dt + c \\
 &= \int dtE[g(t)] + c \\
 &= tE[g(t)] + c \\
\end{align}

In summary, we see that we can only approximate $g(t)$ with its mean value when it's either constant or the system is a simple integrator.

## Do not use this method to find the firing rate with mean feedback approximation

One could try finding the firing rate of the adaptive LIF neuron with mean feedback approximation using an iterative feedback approach. For example,

    def aLIF_firing_rate(input, alpha_f):
        While not converged
            firing rate = LIF_firing_rate(input)
            input -= alpha_f * firing_rate
        return firing_rate

In words, we set the firing rate of the neuron to the firing rate of the LIF neuron given the same input, calculate the amount of feedback, subtract the feedback from the input, and repeat until convergence.

This method fails when the feedback is greater than the input, which is liable to happen when the input is near threshold. When the feedback is greater than the input, the firing rate will converges to $0$ in a single iteration and the algorithm terminates. However, the actual firing rate is not $0$. To illustrate, consider the following plots.

```python
tau_m = .01
tau_f = .005
tref = .005
xt = 1.
af = .02

th_max_f = 1./tref  # theoretical maximum firing rate
max_u = 5.  # maximum u_in to consider
max_f = th_lif_fi(max_u, tau_m, tref, xt)
f = np.array([30., 100.])
u = th_lif_if(f, tau_m, tref, xt)  # linear in firing rate

cc = [(float(i)/(len(u)-1), 0, 0) for i in xrange(len(u))]
alpha = .5
fig = figure(figsize=(16,6))

ax = fig.add_subplot(121)
_u = np.array(sort(np.linspace(0, max_u, 100).tolist() + [xt, 1.001*xt, 1.01*xt, 1.1*xt]))
ax.plot(_u, th_lif_fi(_u, tau_m, tref, xt), 'k', linewidth=2, label='nonadaptive LIF')
for idx, u_val in enumerate(u):
    f = th_lif_fi(u_val, tau_m, tref, xt)
    ax.plot([u_val, ax.get_xlim()[1]], [f, f], c=cc[idx])
    ax.plot([u_val, u_val], [0, f], c=cc[idx], ls=':')
ax.set_ylim(0, th_max_f)
ax.set_xlim(0, max_u*1.001)
ax.legend(loc='upper left')
ax.set_xlabel(r'$u_{in}$', fontsize=20)
ax.set_ylabel(r'$f(u_{in})$ (spks / s)', fontsize=20)

f = np.linspace(0, th_max_f, 201)
uf = af*np.exp(-tref/tau_f)*f

ax = fig.add_subplot(122)
for idx, u_val in enumerate(u):
    net_u = u_val - uf
    f_ss = th_lif_fi(net_u, tau_m, tref, xt)
    ax.plot(uf, f_ss, c=cc[idx], label=r'$u_{in}=%.2f$' % u_val)
    uf0 = 0.
    f0 = f_ss[0]
    for i in xrange(3):
        uf1 = f0*(af*np.exp(-tref/tau_f))
        f1 = th_lif_fi(u_val-uf1, tau_m, tref, xt)
        ax.plot([uf0, uf1], [f0, f0], c=cc[idx], ls=':')
        ax.plot([uf1, uf1], [f0, f1], c=cc[idx], ls=':')
        uf0 = uf1
        f0 = f1
    ax.plot([uf1],[f1], 'o', c=cc[idx], alpha=.7)
iter_uf = af*np.exp(-tref/tau_f)*num_af
ax.plot(uf, f, 'b', label=r"$u_f=\alpha_f'f$")
ax.set_xlim(0, max(uf))
ax.legend(loc='upper left', fontsize=16)
ax.set_xlabel(r'$u_f$', fontsize=20)
ax.set_ylabel(r'$f(u_{in},\ u_f)$ (spks / s)', fontsize=20);
```

Here we illustrate the above algorithm for two values of $u_{in}$. 

At $u_{in}=2.54$, we start at the $0$ feedback $(u_f=0)$ point on red curve on the right plot. Here the neuron fires at about $100$ spikes per second.  We then calculate the amount of feedback by finding the corresponding point on the blue curve as traced by the top dotted red line. Following the red dashed line, we take this amount of feedback and recalculate the neuron's firing rate and so forth until the algorithm converges at the correct intersection between the red and blue lines as denoted by the red circle.

At $u_{in}=1.06$, we use the same algorithm and this time follow the dashed black line. Here we see that the amount of feedback at the $u_f=0$ firing rate overwhelms the input and sets the firing rate to $0$ after a single iteration as denoted by the black circle. However, the intersection between the black curve and the blue curve is not $0$, so the algorithm has incorrectly estimated the firing rate using the mean feedback approximation.
