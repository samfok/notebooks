# Canonical synapse description

The goal of the stochastic digital synapse is to approximate the synaptic differential equation

$$
\tau\dot{x} = -x + u
$$

where $x$ is the state of the synapse and $u$ is the spiking input to the synapse. The impulse response of the synapse is 

$$
x(t) = \frac{1}{\tau}e^{-t/\tau}\text{u}(t)
$$

where $\text{u}$ is the unit step function. Spikes are modeled as delta functions, so 

$$
x(t) = \sum_i\frac{1}{\tau}e^{-(t-t_i)/\tau}\text{u}(t-t_i).
$$

The mathematical synapse has two properties that we will seek to approximate with the stochastic digital synapse:
 - That the area of the impulse response is 1
  - As a consequence, given a input spike rate $f$, the synapse has a mean value of $f$.
 - That the synaptic impulse response decays exponentially with time constant $\tau$

# Stochastic digital synapse description

The stochastic digital synapse mimics the probabilistic nature of ion channels in the neuron membrane. An ion channel has 2 states: opened and closed. When open, the ion channel allows ions, and hence a current, to flow through the membrane. Although each channel has only two states, the cumulative effect of many channels in the syanpse is to make the state of the synapse effectively continuous.

The stochastic digital synapse has channels which open when receiving a spike, and close at each time step with some probability. The value of the digital synapse is the sum of the number of open channels scaled to match the dynamics of the canonical synapse.  

The stochastic digital synapse has the following parameters:
 - $\Delta t$ - the timestep of the synapse
 - $D$ - the state of the digital synapse, in units of $h$
 - $h$ - the scale of the digital synapse
 - $k$ - the number of channels
 - $p$ - the probability of an open channel remaining open
 
The user sets $\Delta t$ and $k$. It is our job to determine $p$ and $h$ so that $h\cdot D[n]$ approximates the continuous synapse equation described by $x(t)$.

Each time the synapse receives a spike, $k$ channels open, so

$$
D := D + k
$$

At each $\Delta t$, the $k$ channels each switch with probability $p$ from $1$ to $0$. Since $x$ only decays at each $\Delta t$ timestep, to simplify our notation when considering the decay of $x$, we use $x(t)=x(\lfloor t/\Delta t\rfloor)=x[n]$, where $n=\lfloor t/\Delta t\rfloor$. In other words

$$
x(t)=x(n\Delta t)=x[n] \quad\text{for}\quad n\Delta t \le t < (n+1)\Delta t
$$

We can formulate the two properties of the synapse as requirements for the stochastic digital synapse:
 - The expected area of the impulse response is 1
 - The impulse response at each $\Delta t$ is, on expectation, scaled by $e^{-\Delta t/\tau}$

## Preserving impulse area

Let's start by assuming that spikes only arrive at $\Delta t$ intervals. We'll handle the case when spikes arrive between $\Delta t$s later.

The lifetime of a channel, $T$, describes the number of $\Delta t$ intervals that a channel is open.

\begin{align}
p_{T}(t) &= \text{Pr}(T=t) \\
 &= p^{t-1}(1-p),\quad t\in\{1,2,\ldots\}
\end{align}

so

$$
E[T] = \frac{1}{1-p}.
$$

For, $k$ channels, the area for the impulse response is

$$
\text{Area} = khT\Delta t,
$$

and since we desire

$$
E[\text{Area}] = 1,
$$

we have 

\begin{align}
1 &= E\left[khT\Delta t\right] \\
  &= kh\Delta tE[T] \\
  &= kh\Delta t\frac{1}{1-p} \\
1-p &= kh\Delta t
\end{align}

## Preserving impulse response time course

Given the number of synapse channels open at the current time, $D[n]$, the number of synapse channels open at the next time step, $D[n+1]$, is a the sum of $D[n]$ coin flips with probability of success $p$.

\begin{align}
p_{D[n+1]|D[n]}(d[n+1]) &= \text{Pr}\left(D[n+1]=d[n+1]\big|D[n]\right) \\
  &= 
  \begin{cases}
  \binom{D[n]}{d[n+1]}p^{d[n+1]}(1-p)^{D[n]-d[n+1]} \quad &\text{for }d[n+1]\in\{0,\ldots,D[n]\}\\
  0 \quad &\text{otherwise,}
  \end{cases}
\end{align}

and 

$$
E\left[D[n+1]\bigg|D[n]\right] = D[n]p.
$$

Discretizing the impulse response of the continuous time synapse yields the update equation

$$
x[n+1] = x[n]e^{\Delta t/\tau} \qquad n=0,1,\ldots
$$


To preserve this decay rate in expectation for the stochastic synapse, we desire

$$
E\left[D[n+1]\bigg|D[n]\right] = D[n]e^{-\Delta t/\tau},
$$

so

$$
p=e^{-\Delta t/\tau}.
$$

Since

$$
1-p = kh\Delta t,
$$

$$
h=\frac{1-e^{-\Delta t/\tau}}{k\Delta t}
$$

Relative to time $0$, the number of channels open in the impulse response at a time $n$ is given by the number of channels that are have lifetimes longer than $n$. Therefore, the distribution of $D[n]$ is given by

\begin{align}
p_{D[n]}(d[n]) &= \text{Pr}(D[n]=d[n])\\
 &= 
 \begin{cases}
 \binom{k}{d[n]}(\text{Pr}(T>n))^{d[n]}(\text{Pr}(T\le n))^{k-d[n]} \\
 0
 \end{cases} \\
 &= 
 \begin{cases}
 \binom{k}{d[n]}(p^n)^{d[n]}(1-p^n)^{k-d[n]} \quad &\text{for } d[n]\in\{0,\ldots,k\}\\
 0 \quad &\text{otherwise.}
 \end{cases}
\end{align}

# Simulation of stochastic digital synapse impulse response

What does the stochastic synapse look like?

```python
# Generate data
%pylab inline
nruns = 30
dt = .001
tau = .01
# ks=[1024]
ks=[1, 4, 16, 64, 256, 1024]
nsteps = 10*round(tau/dt)
p = exp(-dt/tau)
t = arange(nsteps)*dt
impulse_ref = exp(-t/tau)/tau
datas = [zeros((nsteps, nruns)) for k in ks]

for k, data in zip(ks, datas):
    h = (1-p)/(k*dt)

    data[0,:] = k
    for r in xrange(nruns):
        i = 0
        while (i < nsteps-1) and (data[i, r] > 0):
            data[i+1, r] = data[i, r] - sum(rand(data[i,r]) > p)
            i += 1
    data *= h
```

```python
# plot impulse responses
figure('impulse response', figsize=(16, 8))
plt_idx = 1
for k, data in zip(ks, datas):
    subplot(2, 3, plt_idx)
    for r in xrange(nruns):
        step(t, data[:, r], where='post')
    plot(t, impulse_ref, 'r', linewidth=2, label='reference synapse')
    if plt_idx == 1:
        legend(loc='upper right')
    plt_idx += 1
    title(r'$k=%d$' % k, fontsize=18)
suptitle('The stochastic synapse impulse response approximates the '
         +'continuous synapse impulse response.', fontsize=20);
```

```python
# plot area of impulse response
figure('impulse response area', figsize=(16,8))
plt_idx = 1
for k, data in zip(ks, datas):
    area = sum(data, axis=0) * dt
    subplot(2, 3, plt_idx)
    hist(area, bins=nruns)
    plt_idx += 1
    axvline(mean(area), ymin=ylim()[0], ymax=ylim()[1], color='r', linewidth=2)
    xlim(-1,3)
    title(r'$k=%d$' % (k), fontsize=18)
suptitle('The stochastic synapse preserves the unit area of the impulse response. ' +
         '%d runs' % nruns, fontsize=20);
```

```python
# fit and plot effective time constant
figure('impulse response tau fit', figsize=(16,12))
plt_idx = 1
for k, data in zip(ks, datas):
    h = p/(k*dt)
    fit_tau = zeros(nruns)
    max_idx = 0
    subplot(4, 3, plt_idx)
    for r in xrange(nruns):
        fit_idx = data[:, r] > 0
        fit_data = log(data[fit_idx, r] * tau) # normalize
        fit_t = t[fit_idx]
        fit_t.shape = (fit_t.shape[0],1)
        if k>1:
            lstsq_data = lstsq(fit_t, fit_data)
            fit_tau[r] = -1./lstsq_data[0]

        step(fit_t, fit_data, where='post')
    
        if fit_data.shape[0] > max_idx:
            max_idx = fit_data.shape[0]
    log_scaled_impulse = log(impulse_ref[:max_idx]*tau)
    plot(t[:max_idx], log_scaled_impulse, 'r', linewidth=2, label='reference synapse')
    title(r'$k=%d$' % (k), fontsize=18)
    
    if k > 1:
        subplot(4, 3, plt_idx+6)
        hist(fit_tau, bins=nruns)
        axvline(mean(fit_tau), ymin=ylim()[0], ymax=ylim()[1], color='r', linewidth=2)
        axvline(tau, ymin=ylim()[0], ymax=ylim()[1], color='r', linestyle=':', linewidth=1)
        xlim(0,.02)
        title(r'$k=%d,\ \tau=%f$' % (k, mean(fit_tau)), fontsize=18)

    plt_idx += 1

suptitle('The stochastic synapse preserves the time constant of the impulse response. %d runs' % nruns, fontsize=20);
```

# Stochastic digital synapse SNR

The synapse will be receiving spikes from a Poisson process with rate $\lambda$.

## Stochastic digital synapse mean

Consider from the perspective of a point in time of the synapse. Each spike that arrives $n_i$ time steps before the current time contributes $D[n_i]$ to the current value of the synapse independent of any other spike. For $N$ spikes that arrived prior to the current time, the value of the synapse is

$$
S = h\sum_{i=1}^{N}D[n_i].
$$

The synapse mean will then be given by

\begin{align}
E[S] &= E\left[h\sum_{i=1}^{N}D[n_i]\right] \\
 &= hE\left[\sum_{i=1}^{N}D[n_i]\right] \\
 &= hE\left[E\left[\sum_{i=1}^{N}D[n_i]\bigg|N\right]\right] \\
 &= hE\left[\sum_{i=1}^{N}E\left[D[n_i]\bigg|N\right]\right] \\
 &= hE\left[\sum_{i=1}^{N}E\left[D[n_i]\right]\right] \\
 &= hE\left[\sum_{i=1}^{N}1\right]E\left[D[n_i]\right] \\
 &= hE[N]\ E[D[n_i]] \\
 &= hE[N]\ E\left[E\left[D[n_i]\bigg|n_i\right]\right] \\
 &= hE[N]\ E\left[kp^{n_i}\right] \\
 &= hkE[N]\ E\left[p^{n_i}\right] \\
\end{align}



To avoid considering all spikes since the beginning of time, we assume that the impulse response is negligible after a certain time. That is, we only consider spikes within $m$ steps (inclusive) of the current time, where $m$ is an arbitrary, positive integer. The higher $m$ is, the further back in time we consider. As $N$ is the number of spikes from a Poisson spike train with rate $\lambda$ received within $(m+1)\Delta t$, $E[N]=\lambda (m+1)\Delta t$, so

\begin{align}
E[S] &= hkE[N]E\left[p^{n_i}\right] \\
 &= hk\lambda (m+1)\Delta tE\left[p^{n_i}\right]
\end{align}

The spike times within a time period are uniform for a Poisson spike train. With our discretization of time, $n_i$ is uniformly distributed on $[0, m]$, so

\begin{align}
E\left[p^{n_i}\right] &= \sum_{n_i=0}^{m}\frac{1}{m+1}p^{n_i} \\
 &= \frac{1}{m+1}\sum_{n_i=0}^{m}p^{n_i} \\
 &= \frac{1}{m+1}\frac{1-p^{m+1}}{1-p} \\
\end{align}

and 

\begin{align}
E[S] &= hk\lambda(m+1)\Delta tE\left[p^{n_i}\right]\\
 &= hk\lambda(m+1)\Delta t\frac{1}{m+1}\frac{1-p^{m+1}}{1-p}\\
 &= hk\lambda \Delta t\frac{1-p^{m+1}}{1-p}.\\
\end{align}

Substituting $1-p=kh\Delta t$,

\begin{align}
E[S] &= hk\lambda \Delta t\frac{1-p^{m+1}}{1-p} \\
 &= \lambda(1-p)\frac{1-p^{m+1}}{1-p} \\
 &= \lambda(1-p^{m+1}). \\
\end{align}

Taking the limit yields

\begin{align}
\lim_{m\rightarrow\infty}E[S] &= \lim_{m\rightarrow\infty}\lambda(1-p^{m+1}) \\
 &= \lambda
\end{align}

as expected.

## Stochastic digital synapse variance

$$
\text{var}(S)=E[S^2]-E[S]^2
$$

We know $E[S]$, so we need to find $E[S^2]$

\begin{align}
E[S^2] &= E\left[\left(h\sum_{i=1}^{N}D[n_i]\right)^2\right] \\
 &= h^2E\left[\left(\sum_{i=1}^{N}D[n_i]\right)^2\right] \\
 &= h^2E\left[(D[n_1]+D[n_2]+\ldots+D[n_N])(D[n_1]+D[n_2]+\ldots+D[n_N])\right] \\ 
 &= h^2E\left[\sum_{i=1}^{N}D[n_i]^2+\sum_{x=1}^{N}\sum_{y=1,\ y\neq x}^{N}D[n_x]D[n_y]\right] \\
 &= h^2\left(E\left[\sum_{i=1}^{N}D[n_i]^2\right]+E\left[\sum_{x=1}^{N}\sum_{y=1,\ y\neq x}^{N}D[n_x]D[n_y]\right]\right) \\
 &= h^2\left(E\left[E\left[\sum_{i=1}^{N}D[n_i]^2\bigg|N\right]\right]+
    E\left[E\left[\sum_{x=1}^{N}\sum_{y=1,\ y\neq x}^{N}D[n_x]D[n_y]\bigg|N\right]\right]\right) \\
 &= h^2\left(E\left[\sum_{i=1}^{N}E\left[D[n_i]^2\bigg|N\right]\right]+
    E\left[\sum_{x=1}^{N}\sum_{y=1,\ y\neq x}^{N}E\left[D[n_x]D[n_y]\bigg|N\right]\right]\right) \\
 &= h^2\left(E\left[\sum_{i=1}^{N}1\right]E\left[D[n_i]^2\bigg|N\right]+
    E\left[\sum_{x=1}^{N}\sum_{y=1,\ y\neq x}^{N}1\right]E\left[D[n_x]D[n_y]\bigg|N\right]\right) \\
 &= h^2\left(E[N]E\left[D[n_i]^2\right] + E\left[N^2-N\right]E[D[n_x]D[n_y]]\right) \\
 &= h^2\left(E[N]E\left[D[n_i]^2\right] + E\left[N^2-N\right]E[D[n_x]]E[D[n_y]]\right) \\
 &= h^2\left(E[N]E\left[E\left[D[n_i]^2\bigg|n_i\right]\right] + 
    \left(E\left[N^2\right]-
    E[N]\right)E\left[E\left[D[n_x]\bigg|n_x\right]\right]E\left[E\left[D[n_y]\bigg|n_y\right]\right]\right) \\
\end{align}

As before, we first assume that we only consider spikes received $m$ since time intervals ago and we assume that spikes arrive as a Poisson process with rate $\lambda$. 

Therefore, $E[N]=\text{var}(N)=\lambda(m+1)\Delta t$, and $E[N^2]=\lambda(m+1)\Delta t+(\lambda(m+1)\Delta t)^2$. 

In addition, $E\left[D[n]\bigg|n\right]=kp^{n}$ and $E\left[D[n]^2\bigg|n\right]=(k^2-k)p^{2n}+kp^n$, so

\begin{align}
E[S^2] &= h^2\left(E[N]E\left[E\left[D[n_i]^2\bigg|n_i\right]\right] + 
    \left(E\left[N^2\right]-
    E[N]\right)E\left[E\left[D[n_x]\bigg|n_x\right]\right]E\left[E\left[D[n_y]\bigg|n_y\right]\right]\right) \\
 &= h^2\left(\lambda(m+1)\Delta tE\left[(k^2-k)p^{2n_i}+kp^{n_i}\right] + 
    \left(\lambda(m+1)\Delta t\right)^2E\left[kp^{n_x}\right]E\left[kp^{n_y}\right]\right) \\
 &= h^2\left(\lambda(m+1)\Delta tE\left[(k^2-k)p^{2n_i}+kp^{n_i}\right] + 
    \left(\lambda(m+1)\Delta t\right)^2\left(\frac{k}{m+1}\frac{1-p^{m+1}}{1-p}\right)^2\right) \\
 &= h^2\left(\lambda(m+1)\Delta t\left(\frac{k^2-k}{m+1}\frac{1-p^{2(m+1)}}{1-p^2}+\frac{k}{m+1}\frac{1-p^{m+1}}{1-p}\right) + 
    \left(\lambda(m+1)\Delta t\right)^2\left(\frac{k}{m+1}\frac{1-p^{m+1}}{1-p}\right)^2\right) \\
 &= h^2\left(\lambda\Delta t(k^2-k)\frac{1-p^{2(m+1)}}{1-p^2}+\lambda\Delta tk\frac{1-p^{m+1}}{1-p} + 
    \left(\lambda\Delta t\right)^2\left(k\frac{1-p^{m+1}}{1-p}\right)^2\right). \\
\end{align}

Substituting $h=\frac{1-p}{k\Delta t}$,

\begin{align}
E[S^2] &= h^2\left(\lambda\Delta t(k^2-k)\frac{1-p^{2(m+1)}}{1-p^2}+\lambda\Delta tk\frac{1-p^{m+1}}{1-p} + 
    \left(\lambda\Delta t\right)^2\left(k\frac{1-p^{m+1}}{1-p}\right)^2\right) \\
 &= \left(\frac{1-p}{k\Delta t}\right)^2\left(\lambda\Delta t(k^2-k)\frac{1-p^{2(m+1)}}{1-p^2}+\lambda\Delta tk\frac{1-p^{m+1}}{1-p} + 
    (\lambda\Delta t)^2\left(k\frac{1-p^{m+1}}{1-p}\right)^2\right) \\
 &= \frac{(1-p)^2\lambda(k^2-k)}{k^2\Delta t}\frac{1-p^{2(m+1)}}{1-p^2}+
    \frac{(1-p)^2\lambda}{k\Delta t}\frac{1-p^{m+1}}{1-p} + 
    \lambda^2\left(1-p^{m+1}\right)^2 \\
 &= \frac{(1-p)^2\lambda(k^2-k)}{k^2\Delta t}\frac{1-p^{2(m+1)}}{1-p^2}+
    \frac{(1-p)\lambda}{k\Delta t}\left(1-p^{m+1}\right) + 
    \lambda^2\left(1-p^{m+1}\right)^2. \\
\end{align}

Taking the limit yields

\begin{align}
\lim_{m\rightarrow\infty}E[S^2] &= 
    \lim_{m\rightarrow\infty}\left(\frac{(1-p)^2\lambda(k^2-k)}{k^2\Delta t}\frac{1-p^{2(m+1)}}{1-p^2}+
    \frac{(1-p)\lambda}{k\Delta t}\left(1-p^{m+1}\right) + 
    \lambda^2\left(1-p^{m+1}\right)^2\right) \\
 &= \frac{(1-p)^2\lambda(k^2-k)}{k^2\Delta t}\frac{1}{1-p^2}+
    \frac{(1-p)\lambda}{k\Delta t} + \lambda^2.
\end{align}

Therefore,

\begin{align}
\text{var}(S) &= E[S^2]-E[S]^2 \\
 &= \frac{(1-p)^2\lambda(k^2-k)}{k^2\Delta t}\frac{1}{1-p^2}+
    \frac{(1-p)\lambda}{k\Delta t} + \lambda^2 - \lambda^2 \\
 &= \frac{(1-p)^2\lambda(k^2-k)}{k^2\Delta t}\frac{1}{1-p^2}+
    \frac{(1-p)\lambda}{k\Delta t} \\
 &= \frac{(1-p)^2\lambda(k^2-k)}{k^2\Delta t}\frac{1}{(1-p)(1+p)}+
    \frac{(1-p)\lambda}{k\Delta t} \\
 &= \frac{(1-p)\lambda(k^2-k)}{k^2\Delta t}\frac{1}{(1+p)}+
    \frac{(1-p)\lambda}{k\Delta t} \\
 &= \frac{\lambda}{k}\left(\frac{(1-p)}{(1+p)}\frac{(k-1)}{\Delta t}+
    \frac{(1-p)}{\Delta t}\right) \\
\end{align}

Since $p=e^{-\Delta t/\tau}$, we can approximate $p\approx 1-\frac{\Delta t}{\tau}$

\begin{align}
\text{var}(S) &= \frac{\lambda}{k}\left(\frac{(1-p)}{(1+p)}\frac{(k-1)}{\Delta t}+
    \frac{(1-p)}{\Delta t}\right) \\
 &\approx \frac{\lambda}{k}\left(\frac{\Delta t/\tau}{(1+p)}\frac{(k-1)}{\Delta t}+
    \frac{\Delta t/\tau}{\Delta t}\right) \\
 &= \frac{\lambda}{k\tau}\left(\frac{k-1}{1+p}+ 1\right) \\
 &= \frac{\lambda}{k\tau}\left(\frac{k+p}{1+p}\right) \\
 &= \frac{\lambda}{\tau}\left(\frac{1+p/k}{1+p}\right) \\
\end{align}

To build some intution, when $\Delta t\rightarrow 0$, $p\rightarrow 1$ and 

$$
\text{var}(S) = \frac{\lambda}{\tau}\left(\frac{1+1/k}{2}\right),
$$

and now if $k\rightarrow \infty$ as well,

$$
\text{var}(S) = \frac{\lambda}{2\tau},
$$

which matches the result for the continuous synapse. In terms of the original user parameters,

$$
\text{var}(S) = \frac{\lambda}{\tau}\left(\frac{1+e^{-\Delta t/\tau}/k}{1+e^{-\Delta t/\tau}}\right)
$$
